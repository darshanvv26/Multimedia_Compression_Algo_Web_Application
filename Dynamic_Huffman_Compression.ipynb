## Dynamic Huffman Compression Explained

Dynamic Huffman Compression, also known as Adaptive Huffman Coding, is a form of Huffman coding that allows the Huffman tree to adapt to changing symbol frequencies in the data stream. Unlike **Static Huffman Compression**, where the symbol frequencies are pre-calculated from the entire dataset before encoding begins and the tree remains fixed, Dynamic Huffman builds and updates its Huffman tree incrementally as it processes the data. This adaptive approach is crucial when the frequency distribution of symbols is unknown beforehand or changes significantly over time, such as in real-time data streams.

### Advantages over Static Huffman Compression

The primary advantage of Dynamic Huffman Compression lies in its adaptability. In scenarios where the complete data is not available beforehand (e.g., real-time streaming data, network packets), static Huffman is impractical because it requires a priori knowledge of all symbol frequencies to construct the optimal tree. Dynamic Huffman overcomes this limitation by:

*   **Handling Unknown Frequencies:** It can start compressing data without any initial knowledge of symbol frequencies, building the tree as symbols are encountered.
*   **Adapting to Changing Frequencies:** If the frequency distribution of symbols changes during the data stream (e.g., in a multimedia file where different segments have different characteristics), the Huffman tree can be updated to reflect these new frequencies, maintaining compression efficiency.
*   **Single-Pass Encoding/Decoding:** Both the encoder and decoder can build and update the Huffman tree simultaneously using the same algorithm. This eliminates the need to transmit the Huffman tree separately, which is often required in static Huffman, saving overhead.
*   **Improved Efficiency for Dynamic Data:** For streams where symbol probabilities vary over time, a dynamically updating tree can achieve better compression ratios than a static tree built on an average frequency distribution.

### The 'NYT' (Not Yet Transmitted) Node

Central to Dynamic Huffman's adaptability is the concept of the 'NYT' (Not Yet Transmitted) node, sometimes called the '0-node' or 'placeholder node'. This special node plays a crucial role in handling symbols that have not been encountered before in the data stream.

When a new, unseen symbol is encountered during the encoding process:

1.  **Code for NYT:** The encoder first transmits the current Huffman code for the NYT node.
2.  **Code for New Symbol:** Immediately after, it transmits the raw, uncompressed code (e.g., ASCII or a fixed-length representation) for the new symbol.
3.  **Tree Update:** The Huffman tree is then updated. The NYT node is replaced by a parent node, which has two children: a new NYT node and a leaf node representing the newly encountered symbol. The frequency of the new symbol is initialized to 1, and the frequencies of all affected ancestor nodes are incremented. The tree is then re-balanced according to Huffman's rules (e.g., sibling property).

The decoder, operating synchronously, performs the exact same steps. When it decodes the NYT code, it knows the next set of bits will represent a new symbol. It reads this symbol, updates its own tree identically to the encoder, and proceeds. This mechanism allows both encoder and decoder to maintain identical Huffman trees without explicit communication of tree structure, adapting as new symbols appear in the stream.

### Efficiency and Applications in Multimedia Analytics

Dynamic Huffman Compression, while generally offering good compression ratios, does incur some overhead due to the continuous updating and re-balancing of the Huffman tree. This computational cost can be higher than static Huffman, which performs tree construction once. However, its adaptive nature makes it invaluable for specific applications, especially within multimedia analytics and real-time data streams where static methods fall short.

**Key applications include:**

*   **Lossless Data Compression of Streams:** For live audio/video feeds, sensor data, or network traffic, where data arrives sequentially and its statistical properties might change over time, Dynamic Huffman ensures efficient compression without prior knowledge of the entire dataset. Both encoder and decoder can independently build the same Huffman tree, eliminating the need to transmit the tree structure.
*   **Multimedia Analytics:** In analyzing large multimedia datasets (e.g., video surveillance, medical imaging), different segments or frames might have varying statistical properties. Dynamic Huffman can adapt to these changes, potentially yielding better overall compression than a static approach that averages characteristics across the entire dataset. This is crucial for efficient storage and transmission of raw or semi-processed multimedia data.
*   **Adaptive Communication Systems:** In environments with fluctuating channel conditions or varying data types, dynamic compression schemes can adjust to optimize data throughput and reduce latency.
*   **File Formats:** Some file formats or protocols that handle data streams or require adaptive compression might incorporate variants of Dynamic Huffman.

### 1. Initial State of the Dynamic Huffman Tree

Initially, the Dynamic Huffman tree begins with a single special node known as the **NYT (Not Yet Transmitted)** node. This node represents all symbols that have not yet been encountered in the input stream. It has an initial frequency of **0** and serves as a placeholder for new symbols. This NYT node is the root and the only node in the tree at the very beginning.

### 2. Encountering a New Symbol

When a new symbol `S` (one that has not been seen before) is encountered by the encoder:

1.  **Output:** The encoder first outputs the code for the current **NYT (Not Yet Transmitted)** node. Following this, it outputs the fixed-length (e.g., 8-bit ASCII) representation of the new symbol `S` itself.
2.  **Tree Update:**
    *   The existing NYT node is replaced by a new internal parent node.
    *   This new internal parent node will have two children:
        *   A leaf node representing the new symbol `S`, with an initial frequency of **1**.
        *   A new **NYT** node, which will have a frequency of **0**.
3.  **Frequency Increment:** The frequency of the new leaf node (`S`) is incremented to 1. Then, the frequencies of all ancestor nodes, including the new internal parent node, up to the root, are also incremented by 1. This process may trigger a re-balancing of the tree to maintain the Huffman properties (sibling property and Huffman property).

### 3. Encountering an Existing Symbol

When an existing symbol `S` (one that has been seen before) is encountered by the encoder:

1.  **Output:** The encoder outputs the Huffman code corresponding to the leaf node of symbol `S` based on the current state of the Huffman tree.
2.  **Frequency Increment:** The frequency of the leaf node representing symbol `S` is incremented by 1. Subsequently, the frequencies of all its ancestor nodes, all the way up to the root of the tree, are also incremented by 1.
3.  **Tree Re-balancing:** After incrementing frequencies, the tree must be checked and potentially re-balanced to maintain the Huffman properties (sibling property and Huffman property). This process typically involves traversing from the updated leaf node towards the root, identifying any violations, and performing necessary swaps or adjustments.

### 4. Tree Re-balancing (Sibling Property and Huffman Property)

After any frequency increment (whether for a new or existing symbol), the Huffman tree must be re-balanced to maintain two crucial properties:

1.  **Sibling Property:** All nodes are ordered by increasing frequency. Furthermore, for any parent node, its right child must immediately follow its left child in this ordered sequence. This ensures that nodes with lower frequencies are always to the left of nodes with higher frequencies.
2.  **Huffman Property (or Minimum Redundancy Property):** Nodes with higher frequencies are closer to the root of the tree, resulting in shorter codes for more frequent symbols. This is implicitly maintained by ensuring the sibling property and correctly updating frequencies.

**Re-balancing Mechanism:**

*   **Sliding and Swapping:** When a node's frequency increases, it might violate the sibling property. To correct this, the algorithm traverses upwards from the updated node towards the root. If a node's updated frequency is greater than that of a node that should precede it (according to the frequency order), a "slide-and-swap" operation is performed.
    *   The updated node is compared with its 'block leader' (the node with the highest frequency among all nodes with the same frequency). If the updated node's frequency surpasses that of its block leader, it is swapped with the block leader.
    *   This process continues iteratively up the tree until the sibling property is restored, or the root is reached.
*   **Maintaining Tree Structure:** During swaps, the pointers and parent-child relationships within the tree must be carefully updated to preserve the tree's integrity. The goal is to ensure that the node with the increased frequency moves to its correct position in the frequency-ordered sequence of nodes, while maintaining its original siblings and children relative to its new position.

### 5. Encoding Process

When encoding a symbol (either new or existing), the encoder performs the following steps:

1.  **Symbol Lookup:** Locate the leaf node corresponding to the symbol `S` in the current Huffman tree.
2.  **Code Generation:** Traverse from the leaf node `S` upwards to the root of the tree. Each time a left child is encountered, append a `0` to the beginning of the code; each time a right child is encountered, append a `1`. The sequence of `0`s and `1`s from the root to the leaf forms the Huffman code for that symbol.
3.  **Output:** Transmit this generated bit sequence as the encoded representation of the symbol. If the symbol was new, the NYT code and its fixed-length representation would have been sent first, as described in step 2.
4.  **Tree Update:** Immediately after encoding and transmitting, the encoder updates its Huffman tree by incrementing frequencies and re-balancing, as described in steps 2, 3, and 4. This ensures that the tree used for the next symbol is current.

### 6. Decoding Process

The decoder operates in parallel with the encoder, maintaining an identical Huffman tree that is updated synchronously. When the decoder receives an encoded bit stream, it performs the following:

1.  **Bit Reading:** The decoder reads bits from the input stream.
2.  **Tree Traversal:** Starting from the root of its current Huffman tree, the decoder traverses down the tree based on the received bits. A `0` indicates a left child traversal, and a `1` indicates a right child traversal.
3.  **Symbol Identification:**
    *   If the traversal leads to a leaf node that represents a known symbol, that symbol `S` is identified as decoded.
    *   If the traversal leads to the **NYT** node, it means a new symbol is about to be received. The decoder then reads the fixed-length representation (e.g., 8-bit ASCII) that immediately follows the NYT code to identify the new symbol `S`.
4.  **Tree Update:** After identifying a symbol (whether new or existing), the decoder performs the **exact same tree update operations** as the encoder. This includes incrementing frequencies of the identified symbol and its ancestors, and then re-balancing the tree according to the sibling and Huffman properties (as described in steps 2, 3, and 4).
5.  **Output:** The decoded symbol `S` is added to the output stream.

## Dynamic Huffman Compression Example: Input Sequence

Let's walk through the dynamic Huffman compression process with the input sequence: `'ABCA'`.

Initially, the Huffman tree is empty, containing only a Null Symbol (NYT) node, which is the root of the tree.

### Processing Symbol 'A'

*   **Symbol being processed:** `A`

*   **Action (Encoder):**
    *   `A` is a new symbol. The initial tree only contains the NYT node, so the code for NYT is transmitted, followed by the fixed-length raw representation of `A`.
    *   Assuming an 8-bit ASCII representation for `A`, and a '0' code for NYT (as it's the root):
        *   NYT code: `0`
        *   Raw `A`: `01000001` (ASCII for 'A')
    *   **Bits added to total encoded output:** `0` + `01000001`

*   **Tree Update:**
    *   The NYT node is replaced by a new internal parent node. This parent node has two children: a leaf node for `A` with frequency 1, and a new NYT node. The frequency of the parent node is 1.

*   **Tree Re-balancing:**
    *   No re-balancing is needed at this stage as it's the very first symbol, and the tree structure inherently maintains the sibling property (leaf 'A' and NYT are siblings under the new parent).

*   **Current Tree Structure (conceptual):**
    ```
          Root (Weight 1)
         /   \
        A (Weight 1)  NYT (Weight 0)
    ```

*   **Current Huffman Codes:**
    *   A: `0` (derived from the single path from root to A)
    *   NYT: `1` (derived from the single path from root to NYT)
    *   *Note: In practice, the code for the newly added leaf is often the path from the new parent to the leaf, and for the NYT, the path from the new parent to the NYT. Since the parent is the root here, these are direct codes.* However, a more strict view is that the root node now has two children, and these children get codes based on their position (e.g., left child gets '0', right child gets '1'). So, `A` gets `0`, `NYT` gets `1`.

*   **Current Total Encoded Output:** `001000001`

### Processing Symbol 'B'

*   **Symbol being processed:** `B`

*   **Action (Encoder):**
    *   `B` is a new symbol. The encoder looks up the current Huffman code for the NYT node, which is `1`.
    *   It transmits the NYT code, followed by the fixed-length raw representation of `B`.
    *   NYT code: `1`
    *   Raw `B`: `01000010` (ASCII for 'B')
    *   **Bits added to total encoded output:** `1` + `01000010`

*   **Tree Update:**
    *   The existing NYT node (which was a child of the `Root`) is replaced by a new internal node (let's call it `Parent_B`).
    *   `Parent_B` has two children: a new leaf node for `B` with frequency 1, and a new NYT node.
    *   Frequencies are updated: `B` (1), `Parent_B` (1). The `Root` node's frequency increments to 2 (from `A`'s 1 and `Parent_B`'s 1).

*   **Tree Re-balancing:**
    *   The tree structure needs to maintain the sibling property (nodes in the same block are ordered by weight, and an internal node must have a higher weight than its children).
    *   Currently, the `Root` has children `A` (weight 1) and `Parent_B` (weight 1). Their order doesn't violate the property.
    *   `Parent_B` has children `B` (weight 1) and `NYT` (weight 0). This order is also correct.
    *   No re-balancing (node swaps) is necessary at this stage.

*   **Current Tree Structure (conceptual):**
    ```
              Root (Weight 2)
             /           \
            A (Weight 1)  Parent_B (Weight 1)
                         /       \
                        B (Weight 1) NYT (Weight 0)
    ```

*   **Current Huffman Codes:**
    *   A: `0`
    *   B: `10`
    *   NYT: `11`

*   **Current Total Encoded Output:** `001000001101000010`

### Processing Symbol 'C'

*   **Symbol being processed:** `C`

*   **Action (Encoder):**
    *   `C` is a new symbol. The encoder looks up the current Huffman code for the NYT node, which is `11`.
    *   It transmits the NYT code, followed by the fixed-length raw representation of `C`.
    *   NYT code: `11`
    *   Raw `C`: `01000011` (ASCII for 'C')
    *   **Bits added to total encoded output:** `11` + `01000011`

*   **Tree Update:**
    *   The existing NYT node (which was a child of `Parent_B`) is replaced by a new internal node (let's call it `Parent_C`).
    *   `Parent_C` has two children: a new leaf node for `C` with frequency 1, and a new NYT node.
    *   Frequencies are updated: `C` (1), `Parent_C` (1). `Parent_B`'s frequency increments to 2 (from `B`'s 1 and `Parent_C`'s 1). The `Root` node's frequency increments to 3 (from `A`'s 1 and `Parent_B`'s 2).

*   **Tree Re-balancing:**
    *   After updating frequencies, we need to check the sibling property. Nodes in the same block (nodes of the same weight) must be ordered correctly, and weights must propagate correctly.
    *   The `Root` has children `A` (weight 1) and `Parent_B` (weight 2).
    *   `Parent_B` has children `B` (weight 1) and `Parent_C` (weight 1).
    *   `Parent_C` has children `C` (weight 1) and `NYT` (weight 0).
    *   No re-balancing (node swaps) is necessary at this stage as the weights maintain the order and structure.

*   **Current Tree Structure (conceptual):**
    ```
               Root (Weight 3)
              /             \
             A (Weight 1)  Parent_B (Weight 2)
                          /         \
                         B (Weight 1) Parent_C (Weight 1)
                                     /         \
                                    C (Weight 1) NYT (Weight 0)
    ```

*   **Current Huffman Codes:**
    *   A: `0`
    *   B: `10`
    *   C: `110`
    *   NYT: `111`

*   **Current Total Encoded Output:** `0010000011010000101101000011`

### Processing Symbol 'A' (second occurrence)

*   **Symbol being processed:** `A`

*   **Action (Encoder):**
    *   `A` is an existing symbol. The encoder looks up its current Huffman code, which is `0`.
    *   It transmits this Huffman code.
    *   **Bits added to total encoded output:** `0`

*   **Tree Update:**
    *   The frequency of leaf `A` increments to 2.
    *   The frequency of `Root` (ancestor of `A`) increments to 4.

*   **Tree Re-balancing:**
    *   The crucial aspect here is the sibling property and ordering by weight. After `A`'s weight increases to 2, `Root`'s children are `A` (weight 2) and `Parent_B` (weight 2).
    *   According to Huffman properties, nodes of the same weight must be ordered consistently. If the tree implementation dictates that nodes with higher weights should be on the right (or maintain some specific order), a swap might be necessary if `A` was originally on the left and `Parent_B` on the right, and their weights were equal but now `A`'s weight increased.
    *   Let's assume the ADAPTIVE Huffman algorithm's property that nodes of the same weight are ordered by their node number (e.g., higher node number on the left, or vice versa, to maintain canonical order if needed). In a simple weight-based ordering, if `A` (weight 2) and `Parent_B` (weight 2) are siblings, their relative order might need to be maintained or swapped to ensure stability or adherence to the sibling property within the block of weight 2 nodes.
    *   However, a more direct rebalancing rule for Adaptive Huffman is that *nodes must be ordered by weight, from smallest to largest, within each block of siblings*. And *parents must have greater weight than their children*. When a node's weight increases, it might need to move to a 'higher' block of nodes. In our case, `A`'s weight increases to 2, making it equal to `Parent_B`'s weight. If `A` is node `N1` and `Parent_B` is node `N2`, and `N1 < N2` (meaning `A` was added earlier or has a lower node ID), they might need to maintain relative order. More generally, we look for the largest block of nodes with weights less than or equal to `A`'s new weight, and `A` potentially swaps with the largest node in that block if it's not already in the correct position. In this simple example, `A` and `Parent_B` are at the same level and have the same weight. No swap is immediately obvious or strictly needed for this basic illustration, as long as `Parent_B` is still on the right.
    *   Let's assume for simplicity no swap is needed between `A` and `Parent_B` yet, as their weights are now equal and they are direct children of the root.

*   **Current Tree Structure (conceptual):**
    ```
               Root (Weight 4)
              /             \
             A (Weight 2)  Parent_B (Weight 2)
                          /         \
                         B (Weight 1) Parent_C (Weight 1)
                                     /         \
                                    C (Weight 1) NYT (Weight 0)
    ```

*   **Current Huffman Codes:**
    *   A: `0`
    *   B: `10`
    *   C: `110`
    *   NYT: `111`

*   **Current Total Encoded Output:** `00100000110100001011010000110`

````python
import heapq

# Helper functions for binary conversion (Instruction 8)
def char_to_binary(char, bits=8):
    """Converts a character to its fixed-length binary string representation."""
    return bin(ord(char))[2:].zfill(bits)

def binary_to_char(binary_string):
    """Converts a binary string back to a character."""
    return chr(int(binary_string, 2))


class Node:
    """Represents a node in the Huffman tree (Instruction 1)."""
    def __init__(self, symbol=None, frequency=0, node_id=0, left=None, right=None, parent=None):
        self.symbol = symbol  # Character for leaf nodes, None for internal nodes
        self.frequency = frequency
        self.node_id = node_id  # Unique identifier, higher for older nodes
        self.left = left
        self.right = right
        self.parent = parent

    def is_leaf(self):
        return self.left is None and self.right is None

    # For heapq, comparison based on frequency then node_id (lower node_id is older)
    def __lt__(self, other):
        if self.frequency != other.frequency:
            return self.frequency < other.frequency
        return self.node_id < other.node_id

    def __repr__(self):
        return f"Node(id={self.node_id}, symbol={self.symbol!r}, freq={self.frequency}, parent_id={self.parent.node_id if self.parent else 'None'})"


class DynamicHuffmanTree:
    """Manages the Dynamic Huffman Tree (Instruction 2)."""
    def __init__(self, fixed_char_bits=8):
        self.fixed_char_bits = fixed_char_bits
        self.node_counter = 0  # To assign unique node_ids, starting from 0 for NYT
        # Initialize with a single NYT node (Instruction 2)
        self.nyt_node = Node(symbol=None, frequency=0, node_id=self.node_counter)
        self.node_counter += 1
        self.root = self.nyt_node
        self.symbol_nodes = {}  # Stores references to leaf nodes for symbols
        self.nodes_by_id = {self.nyt_node.node_id: self.nyt_node}
        print("DynamicHuffmanTree initialized with NYT node.")

    def _find_node_by_symbol(self, symbol):
        """Searches the tree for a leaf node corresponding to a given symbol (Instruction 3)."""
        return self.symbol_nodes.get(symbol)

    def _get_code(self, node):
        """Returns the Huffman code for a given node as a bit string (Instruction 4)."""
        if node == self.root:
            return ""
        code = []
        current = node
        while current != self.root:
            parent = current.parent
            if parent.left == current:
                code.append('0')
            elif parent.right == current:
                code.append('1')
            current = parent
        return "".join(reversed(code))
````

````python
    def _swap_nodes_in_parent(self, parent_node, child1, child2):
        """Helper to swap two children of a parent node."""
        if parent_node.left == child1 and parent_node.right == child2:
            parent_node.left, parent_node.right = child2, child1
        elif parent_node.left == child2 and parent_node.right == child1:
            parent_node.left, parent_node.right = child1, child2
        # No need to update child.parent, they remain children of parent_node

    def _balance_tree(self, start_node):
        """Re-balances the tree to maintain Huffman properties, traversing up from start_node."""
        current = start_node
        while current:
            # Ensure sibling property: left child's frequency <= right child's frequency
            if current.parent:
                parent = current.parent
                if parent.left and parent.right:
                    if parent.left.frequency > parent.right.frequency:
                        # Swap children if left is heavier than right
                        self._swap_nodes_in_parent(parent, parent.left, parent.right)

            # This simple re-balancing ensures local sibling order. A full Vitter's rebalancing
            # (finding block leaders and performing global swaps) is more complex and usually
            # requires additional data structures like sorted lists of nodes or frequency blocks.
            # For this implementation, we focus on the basic sibling property as nodes are incremented.

            current = current.parent

    def _update_tree(self, symbol):
        """Updates the Huffman tree based on a new or existing symbol (Instruction 5)."""
        node_to_process = self._find_node_by_symbol(symbol)

        if not node_to_process:  # New symbol
            # 1. Get current NYT node
            old_nyt_node = self.nyt_node
            parent_of_nyt = old_nyt_node.parent

            # 2. Create new symbol leaf node and new NYT node
            new_symbol_node = Node(symbol=symbol, frequency=0, node_id=self.node_counter)
            self.node_counter += 1
            new_nyt_node = Node(symbol=None, frequency=0, node_id=self.node_counter)
            self.node_counter += 1

            # 3. Replace old NYT node with an internal node, making new_symbol_node and new_nyt_node its children
            # The old NYT node itself becomes the new internal node
            old_nyt_node.symbol = None  # It's no longer an NYT leaf
            old_nyt_node.left = new_nyt_node # New NYT is typically left child by convention in some FGK/Vitter variants for higher IDs
            old_nyt_node.right = new_symbol_node

            new_nyt_node.parent = old_nyt_node
            new_symbol_node.parent = old_nyt_node

            # Update references
            self.nyt_node = new_nyt_node
            self.symbol_nodes[symbol] = new_symbol_node
            self.nodes_by_id[new_symbol_node.node_id] = new_symbol_node
            self.nodes_by_id[new_nyt_node.node_id] = new_nyt_node

            # Node to start frequency increment from is the new symbol leaf
            node_to_inc_freq = new_symbol_node

        else:  # Existing symbol
            node_to_inc_freq = node_to_process

        # Increment frequencies from the target node up to the root
        current = node_to_inc_freq
        while current:
            current.frequency += 1
            current = current.parent

        # Re-balance the tree. We start from the node whose frequency just incremented.
        # Note: The actual rebalancing for new symbols starts from the new internal node, not the leaf.
        # However, for simplicity, we start from the leaf/updated node and let the _balance_tree logic traverse up.
        if not node_to_process: # If it was a new symbol, rebalance from its parent (the new internal node)
             self._balance_tree(old_nyt_node) # old_nyt_node is now the new internal node
        else: # Existing symbol, rebalance from the symbol's leaf node
            self._balance_tree(node_to_inc_freq)
````

````python
    def encode(self, input_string):
        """Encodes an input string using Dynamic Huffman Compression (Instruction 6)."""
        encoded_bits = []
        for char in input_string:
            node = self._find_node_by_symbol(char)
            if not node:  # New symbol
                # Transmit NYT code + fixed-length raw char
                nyt_code = self._get_code(self.nyt_node)
                encoded_bits.append(nyt_code)
                encoded_bits.append(char_to_binary(char, self.fixed_char_bits))
            else:  # Existing symbol
                encoded_bits.append(self._get_code(node))

            self._update_tree(char) # Update tree for the current character

        return "".join(encoded_bits)
````

````python
    def decode(self, encoded_bits):
        """Decodes an encoded bit string using Dynamic Huffman Compression (Instruction 7)."""
        decoded_output = []
        bit_index = 0

        # Create a new, identical tree for decoding to mirror the encoding process
        # This assumes the decoder starts with the same initial NYT state as the encoder.
        # In a real scenario, this would be a separate instance of DynamicHuffmanTree.
        decoder_tree = DynamicHuffmanTree(fixed_char_bits=self.fixed_char_bits)

        while bit_index < len(encoded_bits):
            current_node = decoder_tree.root
            # Traverse the tree based on the incoming bits until a leaf is reached
            while not current_node.is_leaf():
                if bit_index >= len(encoded_bits):
                    raise ValueError("Premature end of encoded bit stream during tree traversal.")

                bit = encoded_bits[bit_index]
                bit_index += 1

                if bit == '0':
                    current_node = current_node.left
                elif bit == '1':
                    current_node = current_node.right
                else:
                    raise ValueError(f"Invalid bit encountered: {bit}")

            # Symbol identified (leaf node reached)
            if current_node == decoder_tree.nyt_node: # It's the NYT node, so a new symbol follows
                if bit_index + decoder_tree.fixed_char_bits > len(encoded_bits):
                    raise ValueError("Premature end of encoded bit stream while reading new symbol.")

                # Read the fixed-length raw binary representation of the new symbol
                raw_symbol_binary = encoded_bits[bit_index : bit_index + decoder_tree.fixed_char_bits]
                bit_index += decoder_tree.fixed_char_bits

                new_char = binary_to_char(raw_symbol_binary)
                decoded_output.append(new_char)

                # Update decoder's tree with the newly found character
                decoder_tree._update_tree(new_char)

            else: # It's an existing symbol leaf node
                decoded_output.append(current_node.symbol)

                # Update decoder's tree with the existing character
                decoder_tree._update_tree(current_node.symbol)

        return "".join(decoded_output)
````

````python
import heapq

# Helper functions for binary conversion (Instruction 8)
def char_to_binary(char, bits=8):
    """Converts a character to its fixed-length binary string representation."""
    return bin(ord(char))[2:].zfill(bits)

def binary_to_char(binary_string):
    """Converts a binary string back to a character."""
    return chr(int(binary_string, 2))


class Node:
    """Represents a node in the Huffman tree (Instruction 1)."""
    def __init__(self, symbol=None, frequency=0, node_id=0, left=None, right=None, parent=None):
        self.symbol = symbol  # Character for leaf nodes, None for internal nodes
        self.frequency = frequency
        self.node_id = node_id  # Unique identifier, higher for older nodes
        self.left = left
        self.right = right
        self.parent = parent

    def is_leaf(self):
        return self.left is None and self.right is None

    # For heapq, comparison based on frequency then node_id (lower node_id is older)
    def __lt__(self, other):
        if self.frequency != other.frequency:
            return self.frequency < other.frequency
        return self.node_id < other.node_id

    def __repr__(self):
        return f"Node(id={self.node_id}, symbol={self.symbol!r}, freq={self.frequency}, parent_id={self.parent.node_id if self.parent else 'None'})"


class DynamicHuffmanTree:
    """Manages the Dynamic Huffman Tree (Instruction 2)."""
    def __init__(self, fixed_char_bits=8):
        self.fixed_char_bits = fixed_char_bits
        self.node_counter = 0  # To assign unique node_ids, starting from 0 for NYT
        # Initialize with a single NYT node (Instruction 2)
        self.nyt_node = Node(symbol=None, frequency=0, node_id=self.node_counter)
        self.node_counter += 1
        self.root = self.nyt_node
        self.symbol_nodes = {}  # Stores references to leaf nodes for symbols
        self.nodes_by_id = {self.nyt_node.node_id: self.nyt_node}
        print("DynamicHuffmanTree initialized with NYT node.")

    def _find_node_by_symbol(self, symbol):
        """Searches the tree for a leaf node corresponding to a given symbol (Instruction 3)."""
        return self.symbol_nodes.get(symbol)

    def _get_code(self, node):
        """Returns the Huffman code for a given node as a bit string (Instruction 4)."""
        if node == self.root:
            return ""
        code = []
        current = node
        while current != self.root:
            parent = current.parent
            if parent.left == current:
                code.append('0')
            elif parent.right == current:
                code.append('1')
            current = parent
        return "".join(reversed(code))

    def _swap_nodes_in_parent(self, parent_node, child1, child2):
        """Helper to swap two children of a parent node."""
        if parent_node.left == child1 and parent_node.right == child2:
            parent_node.left, parent_node.right = child2, child1
        elif parent_node.left == child2 and parent_node.right == child1:
            parent_node.left, parent_node.right = child1, child2
        # No need to update child.parent, they remain children of parent_node

    def _balance_tree(self, start_node):
        """Re-balances the tree to maintain Huffman properties, traversing up from start_node."""
        current = start_node
        while current:
            # Ensure sibling property: left child's frequency <= right child's frequency
            if current.parent:
                parent = current.parent
                if parent.left and parent.right:
                    if parent.left.frequency > parent.right.frequency:
                        # Swap children if left is heavier than right
                        self._swap_nodes_in_parent(parent, parent.left, parent.right)

            # This simple re-balancing ensures local sibling order. A full Vitter's rebalancing
            # (finding block leaders and performing global swaps) is more complex and usually
            # requires additional data structures like sorted lists of nodes or frequency blocks.
            # For this implementation, we focus on the basic sibling property as nodes are incremented.

            current = current.parent

    def _update_tree(self, symbol):
        """Updates the Huffman tree based on a new or existing symbol (Instruction 5)."""
        node_to_process = self._find_node_by_symbol(symbol)

        if not node_to_process:  # New symbol
            # 1. Get current NYT node
            old_nyt_node = self.nyt_node
            parent_of_nyt = old_nyt_node.parent

            # 2. Create new symbol leaf node and new NYT node
            new_symbol_node = Node(symbol=symbol, frequency=0, node_id=self.node_counter)
            self.node_counter += 1
            new_nyt_node = Node(symbol=None, frequency=0, node_id=self.node_counter)
            self.node_counter += 1

            # 3. Replace old NYT node with an internal node, making new_symbol_node and new_nyt_node its children
            # The old NYT node itself becomes the new internal node
            old_nyt_node.symbol = None  # It's no longer an NYT leaf
            old_nyt_node.left = new_nyt_node # New NYT is typically left child by convention in some FGK/Vitter variants for higher IDs
            old_nyt_node.right = new_symbol_node

            new_nyt_node.parent = old_nyt_node
            new_symbol_node.parent = old_nyt_node

            # Update references
            self.nyt_node = new_nyt_node
            self.symbol_nodes[symbol] = new_symbol_node
            self.nodes_by_id[new_symbol_node.node_id] = new_symbol_node
            self.nodes_by_id[new_nyt_node.node_id] = new_nyt_node

            # Node to start frequency increment from is the new symbol leaf
            node_to_inc_freq = new_symbol_node

        else:  # Existing symbol
            node_to_inc_freq = node_to_process

        # Increment frequencies from the target node up to the root
        current = node_to_inc_freq
        while current:
            current.frequency += 1
            current = current.parent

        # Re-balance the tree. We start from the node whose frequency just incremented.
        # Note: The actual rebalancing for new symbols starts from the new internal node, not the leaf.
        # However, for simplicity, we start from the leaf/updated node and let the _balance_tree logic traverse up.
        if not node_to_process: # If it was a new symbol, rebalance from its parent (the new internal node)
             self._balance_tree(old_nyt_node) # old_nyt_node is now the new internal node
        else: # Existing symbol, rebalance from the symbol's leaf node
            self._balance_tree(node_to_inc_freq)

    def encode(self, input_string):
        """Encodes an input string using Dynamic Huffman Compression (Instruction 6)."""
        encoded_bits = []
        for char in input_string:
            node = self._find_node_by_symbol(char)
            if not node:  # New symbol
                # Transmit NYT code + fixed-length raw char
                nyt_code = self._get_code(self.nyt_node)
                encoded_bits.append(nyt_code)
                encoded_bits.append(char_to_binary(char, self.fixed_char_bits))
            else:  # Existing symbol
                encoded_bits.append(self._get_code(node))

            self._update_tree(char) # Update tree for the current character

        return "".join(encoded_bits)

    def decode(self, encoded_bits):
        """Decodes an encoded bit string using Dynamic Huffman Compression (Instruction 7)."""
        decoded_output = []
        bit_index = 0

        # Create a new, identical tree for decoding to mirror the encoding process
        # This assumes the decoder starts with the same initial NYT state as the encoder.
        # In a real scenario, this would be a separate instance of DynamicHuffmanTree.
        decoder_tree = DynamicHuffmanTree(fixed_char_bits=self.fixed_char_bits)

        while bit_index < len(encoded_bits):
            current_node = decoder_tree.root
            # Traverse the tree based on the incoming bits until a leaf is reached
            while not current_node.is_leaf():
                if bit_index >= len(encoded_bits):
                    raise ValueError("Premature end of encoded bit stream during tree traversal.")

                bit = encoded_bits[bit_index]
                bit_index += 1

                if bit == '0':
                    current_node = current_node.left
                elif bit == '1':
                    current_node = current_node.right
                else:
                    raise ValueError(f"Invalid bit encountered: {bit}")

            # Symbol identified (leaf node reached)
            if current_node == decoder_tree.nyt_node: # It's the NYT node, so a new symbol follows
                if bit_index + decoder_tree.fixed_char_bits > len(encoded_bits):
                    raise ValueError("Premature end of encoded bit stream while reading new symbol.")

                # Read the fixed-length raw binary representation of the new symbol
                raw_symbol_binary = encoded_bits[bit_index : bit_index + decoder_tree.fixed_char_bits]
                bit_index += decoder_tree.fixed_char_bits

                new_char = binary_to_char(raw_symbol_binary)
                decoded_output.append(new_char)

                # Update decoder's tree with the newly found character
                decoder_tree._update_tree(new_char)

            else: # It's an existing symbol leaf node
                decoded_output.append(current_node.symbol)

                # Update decoder's tree with the existing character
                decoder_tree._update_tree(current_node.symbol)

        return "".join(decoded_output)

    def _build_graphviz_tree(self, graph, node):
        """Recursively adds nodes and edges to the graphviz Digraph object."""
        if node is None:
            return

        label = f"ID: {node.node_id}\nFreq: {node.frequency}"
        if node == self.nyt_node:
            label = f"NYT (ID: {node.node_id})\nFreq: {node.frequency}"
        elif node.symbol is not None:
            label = f"'{node.symbol}' (ID: {node.node_id})\nFreq: {node.frequency}"
        else:
            label = f"Internal (ID: {node.node_id})\nFreq: {node.frequency}"

        graph.node(str(node.node_id), label=label)

        if node.left:
            graph.edge(str(node.node_id), str(node.left.node_id), label='0')
            self._build_graphviz_tree(graph, node.left)
        if node.right:
            graph.edge(str(node.node_id), str(node.right.node_id), label='1')
            self._build_graphviz_tree(graph, node.right)

    def visualize_tree(self, filename='dynamic_huffman_tree'):
        """Creates and renders a graphviz visualization of the Huffman tree."""
        dot = graphviz.Digraph(comment='Dynamic Huffman Tree', graph_attr={'rankdir': 'TB'})
        self._build_graphviz_tree(dot, self.root)
        dot.render(filename, view=True, format='png')
        print(f"Tree visualization saved to {filename}.png")

````
